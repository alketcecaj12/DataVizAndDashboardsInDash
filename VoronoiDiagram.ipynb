```python
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Define some random points
points = np.random.rand(10, 2)  # 10 random points in 2D

# Create Voronoi diagram
vor = Voronoi(points)

# Plot Voronoi diagram with custom colors
fig, ax = plt.subplots()

for region_index in range(len(vor.regions)):
    region = vor.regions[region_index]
    if not -1 in region and len(region) > 0:  # Ignore regions that are complex or empty
        polygon = [vor.vertices[i] for i in region]
        ax.fill(*zip(*polygon), alpha=0.4, color=cm.tab20(region_index % 20))

# Plot Voronoi vertices and points
ax.plot(points[:, 0], points[:, 1], 'o', markersize=5, color='black')
ax.plot(vor.vertices[:, 0], vor.vertices[:, 1], 'o', markersize=3, color='orange')

# Show the plot
plt.show()
```

```python
import numpy as np
from scipy.spatial import Voronoi
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def plot_voronoi(points):
    """
    Computes and plots a Voronoi diagram for a set of given points.
    
    Parameters:
    points (array-like): An array of coordinate tuples [(x1, y1), (x2, y2), ...].
    """
    # Create the Voronoi diagram
    vor = Voronoi(points)
    
    # Create a figure and axis
    fig, ax = plt.subplots()
    
    # Plot the Voronoi diagram
    for i, region in enumerate(vor.regions):
        if not -1 in region and len(region) > 0:  # Filter out regions that go to infinity
            polygon = [vor.vertices[i] for i in region]
            ax.fill(*zip(*polygon), alpha=0.4, color=cm.tab20(i % 20))  # Fill region with a color

    # Plot the Voronoi vertices and input points
    ax.plot(points[:, 0], points[:, 1], 'ko', markersize=5)  # Input points in black
    ax.plot(vor.vertices[:, 0], vor.vertices[:, 1], 'ro', markersize=3)  # Vertices in red

    # Set limits and show the plot
    ax.set_xlim(np.min(points[:,0])-0.1, np.max(points[:,0])+0.1)
    ax.set_ylim(np.min(points[:,1])-0.1, np.max(points[:,1])+0.1)
    plt.title("Voronoi Diagram with Colored Regions")
    plt.show()

```

### Example of usage

```python
# Example usage
points = np.array([(0.1, 0.2), (0.4, 0.8), (0.8, 0.3), (0.5, 0.5), (0.3, 0.3), 
                  (0.3, 0.9), (0.7, 0.79), (0.3, 0.39)])
plot_voronoi(points)
```
