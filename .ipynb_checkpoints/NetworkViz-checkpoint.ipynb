### Visualize network connections

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.erdos_renyi_graph(10, 0.5)
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700)
plt.title("Network Graph Example with NetworkX")
plt.show()
```

### Each number is a node connected to other nodes by edges
