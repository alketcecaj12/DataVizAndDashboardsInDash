## Visualize data distrib. using joyplot.

```python
import joypy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'value':np.random.randn(2000), 
                  'month': np.random.choice(['jan', 'feb', 'march', 'apr'], 2000)})
```

```python
joypy.joyplot(df, by= 'month', column = 'value', colormap = plt.cm.coolwarm)
plt.show()
```
