```python
import matplotlib.pyplot as plt
with plt.xkcd(
        scale=4,        # scale level
        length=120,     # scale length
        randomness=2):  #random numer
    plt.figure(dpi=150)
    patches, texts, autotexts = plt.pie(
        x=[1, 2, 3],
        labels=['A', 'B', 'C'],
        colors=['#dc2624', '#2b4750', '#45a0a2'],
        autopct='%.2f%%',
        explode=(0.1, 0, 0))
    texts[1].set_size('20')

    #matplotlib.patches.Wedge
    patches[0].set_alpha(0.3)
    patches[2].set_hatch('|') 
    patches[1].set_hatch('x')
    plt.legend(
        patches,
        ['A', 'B', 'C'],
        title="Pie Learning",
        loc="center left",
        fontsize=15,
        bbox_to_anchor=(1, 0, 0.5, 1))
    plt.title('Lovely pie', size=20)
    plt.show() 
```

The script creates a pie chart using Matplotlib’s xkcd hand‑drawn style. The with plt.xkcd(...) context modifies the entire figure to look like a sketch, adjusting line jitter, length, and randomness. Inside the context the code builds a pie chart with three slices, custom colors, hatching, transparency, an exploded slice, a legend, and a title.

### Grouped barchart

```python
import matplotlib.pyplot as plt
import numpy as np
with plt.xkcd():
    plt.figure(dpi=150)
    labels = ['Jack', 'Rose', 'Jimmy']
    year_2019 = np.arange(1, 4)
    year_2020 = np.arange(1, 4) + 1
    bar_width = 0.4

    plt.bar(
        np.arange(len(labels)) - bar_width / 2,
        year_2019,
        color='#dc2624',
        width=bar_width,
        label='year_2019'
    )
    plt.bar(
        np.arange(len(labels)) + bar_width / 2,
        year_2020,
        color='#45a0a2',
        width=bar_width,
        label='year_2020'
    )
    plt.xticks(np.arange(0, 3, step=1), labels, rotation=45)
    plt.legend(loc=2)
    plt.show()
```
