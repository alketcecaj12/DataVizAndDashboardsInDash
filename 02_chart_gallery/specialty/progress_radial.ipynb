### Progress Radial gauge
The code creates a radial gauge (speedometer‑style) indicator that visualizes the proportion of achieved sales against a target goal. Plotly’s go.Indicator with mode="gauge+number" draws a circular dial, fills a bar up to the current value, and places the numeric value at the center.

```python
import plotly.graph_objects as go

goal = 10000
achieved = 9200

fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=achieved,
    gauge={'axis': {'range':[0,goal]}, 'bar': {'color':'#636EFA'}},
    title={'text': "Annual Sales Progress"}
))
fig.show()
```
