### Radar chart

```python
import numpy as np
import matplotlib.pyplot as plt

def radar_chart(labels, averages, customer_data):
    """ utility function to generate radar chart with range numbers """
    num_vars = len(labels)

    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # Complete the loop
    angles += angles[:1]
    averages += averages[:1]
    customer_data += customer_data[:1]

    # Set up the radar chart
    fig, ax = plt.subplots(figsize=(4, 4), subplot_kw=dict(polar=True))

    ax.fill(angles, averages, color='blue', alpha=0.25)
    ax.plot(angles, averages, color='blue', linewidth=2, label='Average')
    ax.fill(angles, customer_data, color='red', alpha=0.25)
    ax.plot(angles, customer_data, color='red', linewidth=2, label='Customer')

    # Set the range on the y-axis from 0 to 1 with increments of 0.2
    range_values = np.arange(0, 1.1, 0.2)
    ax.set_yticks(range_values)
    ax.set_yticklabels([f'{r:.1f}' for r in range_values])

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    plt.legend(loc='upper right')
    plt.show()

# Example usage
labels = ['A', 'B', 'C', 'D', 'E']
averages = [0.6, 0.7, 0.8, 0.9, 0.85]
customer_data = [0.5, 0.65, 0.75, 0.8, 0.7]
radar_chart(labels, averages, customer_data)
```
